```python
import matplotlib.pyplot as plt 
import numpy as np 
x=np.linspace(0,5,11) 
y = x**2 
x
```

```python
plt.plot(x,y, 'b')
plt.xlabel('X Axis') 
plt.ylabel('Y Axis') 
plt.title('X-Y graph') 
plt.show
```

```python
plt.subplot(1,3,2) 
plt.plot(x,y,'b--') 
plt.subplot(1,3,1)
plt.plot(x,y,'r*-') 
plt.subplot(1,3,3) 
plt.plot(x,y,'g^')
```
